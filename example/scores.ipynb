```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df1 = pd.read_csv("example-data-prog1.csv")
elapsed1 = df1["elapsed"]
scores1 = df1["scores"]

df2 = pd.read_csv("example-data-prog2.csv")
elapsed2 = df2["elapsed"]
scores2 = df2["scores"]

df3 = pd.read_csv("example-data-prog3.csv")
elapsed3 = df3["elapsed"]
scores3 = df3["scores"]
```

```python
fig, ax = plt.subplots()
ax.plot(elapsed1[2::2], scores1[2::2], color="blue", label="Inference Program 1")
ax.plot(elapsed3[2:], scores3[2:], color="green", label="Inference Program 2")
ax.plot(elapsed2[2:], scores2[2:], color="orange", label="Inference Program 3")
ax.legend(loc="lower right")
ax.set_ylabel("Log Probability")
ax.set_xlabel("Runtime (seconds)")
ax.set_xlim((0, 8))

fig.set_tight_layout(True)
fig.set_size_inches((7, 2.5))
fig.savefig("scores.pdf")
```
